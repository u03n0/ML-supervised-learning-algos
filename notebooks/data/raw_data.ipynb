### 1.4 create new df with desired columns

```python
new_df = df[['id', 'about', 'keywords', 'specialities_x', 'Label']]
new_df
```

### 1.5 replace NaN with an empty string in **specialities_x**

```python
new_df['specialities_x'] =  new_df['specialities_x'].fillna('')
new_df
```

### 1.6 remove emojis from **about** and also :smiley_face: which remains after removing emojis

```python
import emoji
import re


def remove_emojis(text: str)-> str:
    """ removes emojis and :emoji: syntax from text.
    """
    text_no_emojis = emoji.demojize(text)
    cleaned_text = re.sub(r':[^:]+:', '', text_no_emojis)
    return cleaned_text


new_df['about'] = df['about'].apply(remove_emojis)
new_df
```

### 1.9 **specialities_x** **keywords** are acutally strings that look like lists. They need to be converted to lists.

```python
from typing import List 


def convert_to_list(text:str)-> List[str]:
    """ Removal of unwanted words and characters from strings.
    Finally, conversion to a list.
    """
    cleaned = text.replace('and', '').replace("[", '').replace("]", '').replace("'", '').split(',')
    return cleaned


target_columns = ['specialities_x', 'keywords']
new_df[target_columns] = new_df[target_columns].applymap(convert_to_list)

new_df
```

### 1.10 Creating a new column as the set of **keywords** and **specialities_x**

```python
new_df['merged'] = new_df.apply(lambda row: list(set(row['keywords'] + row['specialities_x'])), axis=1)
new_df
```

### 1.11 Dropping unnecessary columns 

```python
columns_to_drop = ['keywords', 'specialities_x']
new_df.drop(columns=columns_to_drop, inplace=True)
new_df
```

### 1.12 Changing list into string for **merged**, concatinating **merged** with **about**.
This will add more crucial words to the text we will use for binary classification.
Dropping 'merged' column

```python
new_df['merged'] = new_df['merged'].apply(lambda x: ' '.join(x))
new_df['about'] = new_df['about']  + new_df['merged']
new_df.drop(columns=['merged'], inplace=True)
new_df
```
